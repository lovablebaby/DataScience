```python
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
```

```python
# seaborn setting

sns.set(palette='hls', font_scale=1.5)
```

```python
points = np.random.randn(50).cumsum() # cumsum : 누적 합
points
```

```python
plt.plot(points)
plt.show()
```

```python
plt.plot(range(0,250,5), points) # 1:1 대응이 되도록 지정
plt.show()
```

```python
plt.plot(points,'co:')
plt.show()

# color, marker, line
```

```python
plt.plot(points, 'co-', lw=3, ms=5, mfc='k')
plt.xlim(-30,80)
plt.ylim(-10,20)
# lw = line_wide, ms = marker_size, mfc = color
plt.show()
```

```python
plt.plot(points, 'co-', lw=3, ms=5, mfc='k')
plt.xlim(-30,80)
plt.ylim(-10,20)
plt.xticks([0,25,50])
plt.yticks([-5,0,15],[r'$\theta$a',r'$\theta$b',r'$\theta$c']) # label
plt.show()
```

```python
plt.plot(points, points, 'bo', 
        points, 2*points, 'cs-',
        points, 0.5*points, 'r.', lw=5, ms=10)
plt.show()
```

```python
plt.plot(points, 'co-',
        points*0.5, points, lw=3, ms=5, mfc='b')
plt.show()
```

```python
# legend, title
# 
plt.rc('font', family='nanumgothic') # 범례 한글 설정시 깨지므로, 폰트 설정해야 함
plt.plot(points, label='random points')
plt.plot(0.5*points, label=U'임의값')
plt.legend() # default로 빈 칸 많은 곳으로 감

plt.xlabel('random x')
plt.ylabel('random y')
plt.title('Random plot')

plt.show()
```

```python
plt.plot(points)
 #text, arrow point, xy coordinate,    # text location from text coordinate, text coordinate
plt.annotate(r'(explain)', xy=(40,8), xycoords='data', xytext=(-50,50), textcoords='offset points',
             fontsize=20, arrowprops=dict(arrowstyle='->', linewidth=3, color='b')) # blah blah~
plt.show()
```

```python
plt.figure(figsize=(20,3))
plt.plot(points)
plt.show()
```

```python
# draw several graphs
## Axes, Subplots

ax1 = plt.subplot(2, 1, 1) # 2행 1열의 첫번째 서브플롯
plt.plot(points)

ax2 = plt.subplot(2, 1, 2)
plt.plot(np.random.randn(50))
plt.show()
```

```python
# bar chart

x = [3,2,1]
y = [1,2,3]

# labeling
plt.xticks(x, [U'한개', U'두개', U'세개']) # X에 대응한다.


plt.bar(x,y) # bar 뒤에 h 붙이면, 수평으로 바뀜.
plt.show()
```

```python
# Histogram

x = np.random.randint(0,10,10)
print(x)

arrays, bins, patches = plt.hist(x,bins=6)
plt.show()
```

```python
# pie-chart
# Size, Size-label,
plt.pie([30, 50, 10],
       labels = [U'피자', U'햄버거', U'감자튀김'],
       colors = ['pink', 'salmon', 'tomato'],
       explode = (0,0,0.2), # 띄우는 정도
       autopct = '%.2f%%',  shadow=True, startangle=0)
plt.axis('equal') # 기울이지 않기
plt.title(U'품목별 매출비중')
plt.show()
```
