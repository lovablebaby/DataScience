### Linear Regression with Abalone data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from random import *
```

```python
df = pd.read_csv("Desktop/abalone.csv",header=None, names = ["Length", "Diameter"])
```

```python
print(df.head())
print(df.tail())
```

```python
# X(Input) - Length
Input = np.array(df["Length"])
# Y(Output) - Diameter
Output = np.array(df["Diameter"])
```

### Define functions

```python
def hypothesis(w,b,x):
    h=w*x+b
    return h

def costFunction(Model, Output):
    c = (sum(pow((Model - Output), 2)))/(2*len(Model))
    return c

def GradientDecent(Input, Output, w, a):
    multi_wx = w * Input
    loss = multi_wx - Output
    gradient = np.sum(loss*Input) / len(Input)
    w = w-a*gradient
    return w
```
