### 범례 설정, 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine", color="g")
plt.hold(True)
plt.plot(X, S, label="sine",color='r')
plt.legend(loc=0)
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title('Cosine Plot')
plt.show()
```

## Axes와 Subplot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1)*np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.show
```

```python
plt.subplot(221); plt.plot([1, 2]); plt.title(1)
plt.subplot(222); plt.plot([1, 2]); plt.title(2)
plt.subplot(223); plt.plot([1, 2]); plt.title(3)
plt.subplot(224); plt.plot([1, 2]); plt.title(4)
plt.tight_layout()
plt.show()
```
