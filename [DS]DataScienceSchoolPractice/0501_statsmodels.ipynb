```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
df = sm.datasets.get_rdataset("Guerry", "HistData").data
df.tail()
```

```python
# regression

result = smf.ols('Lottery ~ Literacy + np.log(Pop1831)', data=df).fit()
print(result.summary())
```
